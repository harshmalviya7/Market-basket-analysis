```python
records = []
for i in range(0, 7501):
    records.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
records[0]
```

```python
te_ary = te.fit(records).transform(records)
data = pd.DataFrame(te_ary, columns=te.columns_)
data
```

```python
te=TransactionEncoder()
oht_ary = te.fit(records).transform(records, sparse=True)
sparse_df = pd.DataFrame.sparse.from_spmatrix(oht_ary, columns=te.columns_)
sparse_df.fillna(0,inplace=True)
```

```python
sparse_df.drop(['nan'],axis=1,inplace=True)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets = apriori(sparse_df, min_support = 0.0055, use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric = 'support', min_threshold = 0.0001)

rules.head()
```

```python
rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold = 0.01)

rules
```

```python
rules = rules[rules['consequent support'] > 0.095]
rules
```

```python
rules = rules[rules['leverage'] > 0.0]
rules
```

```python
rules=rules[rules['lift'] > 2]
rules
```

```python
rules.info()
```

```python
rules.head(50)
```

```python
antecedents=[]
for i in rules['antecedents']:
    antecedents.append(list(i))
```

```python
consequents=[]
for i in rules['consequents']:
    consequents.append(list(i))
```

```python
output=set()
def predict(name):
    for count,ele in enumerate(antecedents):
        if name in ele:
            for i in consequents[count]:
                output.add(i)
    return output
```
